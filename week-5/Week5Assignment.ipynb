```python
import pandas as pd

# Load the datasets
students_df = pd.read_csv('Student questionnaire data_REDS_UAE.csv')
teachers_df = pd.read_csv('Teacher questionnaire data_REDS_UAE.csv')
schools_df = pd.read_csv('School head questionnaire data_REDS_UAE.csv')

# Step 1: Aggregate Teacher Data to School Level
# Select relevant teacher columns and aggregate
teacher_agg = teachers_df.groupby('IDSCHOOL').agg({
    'IT06A_teach_qual': 'mean',   # Mean teaching quality
    'IT05F_indiv': 'mean'         # Mean individual assistance
}).reset_index()

# Rename aggregated columns for clarity
teacher_agg.columns = ['IDSCHOOL', 'TeachQual_Mean', 'IndivAssist_Mean']

# Step 2: Merge Datasets
# Merge students with aggregated teacher data
student_teacher_df = pd.merge(students_df, teacher_agg, on='IDSCHOOL', how='left')

# Merge the result with school data
merged_df = pd.merge(student_teacher_df, schools_df, on='IDSCHOOL', how='left')

# Step 3: Check shape of final dataset
print(f"Merged dataset dimensions: {merged_df.shape}")
```

```python
# Save the merged dataset to a CSV file
merged_df.to_csv('merged_REDS_UAE_dataset.csv', index=False)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np

# Define the target and feature columns
target = 'IS18C_plan_capacity'
exclude = ['IDSTUD', 'IDSCHOOL', target]
features = [col for col in merged_df.columns if col not in exclude]

# Prepare the data
X = merged_df[features]
y = merged_df[target]

# Impute missing values with median for numerical columns
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)
y = y.dropna()
X_imputed = X.loc[y.index]  # Align X to y's index
X_imputed = imputer.fit_transform(X_imputed)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=42)

# Train decision tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

# Get feature importances
feature_importances = pd.Series(tree.feature_importances_, index=X.columns).sort_values(ascending=False)
feature_importances.head(5)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
selected_features = [
    'ASDAGE', 'SEX_female', 'prnt_ed_M', 'IS14A', 'IS14K',
    'IS17A', 'IS17C', 'IS21_teach_sup', 'IS09_comm_modes',
    'TeachQual_Mean', 'IndivAssist_Mean'
]

# Keep rows with no missing values in selected features and target
logit_df = merged_df[selected_features + ['IS18C_plan_capacity']].dropna()

# Define input features and target
X = logit_df[selected_features]
y = logit_df['IS18C_plan_capacity']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train multinomial logistic regression
logit_model = LogisticRegression( solver='lbfgs', max_iter=1000)
logit_model.fit(X_scaled, y)

# Predict and get evaluation report
y_pred = logit_model.predict(X_scaled)
report = classification_report(y, y_pred, output_dict=True)

# Get model coefficients
coef_df = pd.DataFrame(logit_model.coef_, columns=selected_features, index=logit_model.classes_)

coef_df, report
```
